```python
def compound_interest(p, r, n, t):
    # A = P(1 + r/n)^(nt)
    amount = p * (1 + r/n)**(n*t)
    return f"Principal: {p}. Rate: {r}. Amount in {t} year(s): ${amount:.2f}"
```

```python
compound_interest(10000, 0.0375, 365, 5)
```

```python
def compound_interest_yty(p, r, n, t):
    # A = P(1 + r/n)^(nt)
    amount_dict = {'Y0':p}
    for i in range(1, t+1):
        amount = p * (1 + r/n)**(n*i)
        amount_dict[f"Y{i}"] = round(amount, 2)
    return amount_dict
```

```python
savings = compound_interest_yty(10000, 0.0375, 365, 30)
savings
```

```python
plt.figure(figsize=(12, 6))
plt.bar(x=savings.keys(), height=savings.values())
plt.xlabel('Year')
plt.ylabel('Future Value')
plt.tight_layout()
plt.show()
```

```python
def yearly_contribution_compound_interest_yty(p, c, r, t):
    """
    Calculate compound interest with yearly contributions.

    Parameters:
    p (float): Initial principal amount
    c (float): Annual contribution
    r (float): Annual interest rate (decimal, e.g., 0.05 for 5%)
    n (int): Number of compounding periods per year
    t (int): Total number of years

    Returns:
    dict: Yearly accumulated values, e.g., {'Y0': 100, 'Y1': ..., ..., 'Yt': ...}
    """
    n=1 # compound once a year
    amount_dict = {'Y0': round(p, 2)}
    for i in range(1, t + 1):
        growth = (1 + r / n) ** (n * i)
        contributions = c * (growth - 1) / (r / n)
        amount = p * growth + contributions
        amount_dict[f"Y{i}"] = round(amount, 2)
    return amount_dict
```

```python
savings_wContributionYearly = yearly_contribution_compound_interest_yty(10000, 1000, 0.0375, 30)
savings_wContributionYearly
```

```python
plt.figure(figsize=(12, 6))
plt.bar(x=savings_wContributionYearly.keys(), height=savings_wContributionYearly.values())
plt.xlabel('Year')
plt.ylabel('Future Value')
plt.tight_layout()
plt.show()
```

```python
def daily_contribution_compound_interest_yty(p, c, r, t):
    """
    Simulate compound interest with yearly contributions but arbitrary compounding frequency (e.g., daily).
    
    p: initial principal
    c: yearly contribution (added end of each year)
    r: annual interest rate (as decimal)
    n: compounding periods per year (e.g., 365 for daily)
    t: total number of years
    """
    n=365
    amount_dict = {'Y0': round(p, 2)}
    current_amount = p

    for i in range(1, t + 1):
        # compound for 1 year
        current_amount = current_amount * (1 + r / n) ** n
        # add yearly contribution
        current_amount += c
        amount_dict[f"Y{i}"] = round(current_amount, 2)

    return amount_dict
```

```python
savings_wContributionDaily = daily_contribution_compound_interest_yty(10000, 1000, 0.0375, 30)
savings_wContributionDaily
```

```python
plt.figure(figsize=(12, 6))
plt.bar(x=savings_wContributionDaily.keys(), height=savings_wContributionDaily.values())
plt.xlabel('Year')
plt.ylabel('Future Value')
plt.tight_layout()
plt.show()
```
